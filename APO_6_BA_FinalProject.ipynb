```python
from sklearn import metrics
num_trees_list = []
training_accuracy_list = []

# Loop through different numbers of trees
for num_trees in range(1, 1000, 50):  # Test from 1 tree to 500 trees with a step of 50
    # Create a Random Forest Regressor with the current number of trees
    rf = RandomForestClassifier(n_estimators=num_trees, random_state=42)

    # Train the model
    rf.fit(X_train, y_train)

    # Make predictions on the test set
    y_pred = rf.predict(X_test)

    # Calculate RMSE (Root Mean Squared Error)
    training_accuracy = metrics.accuracy_score(y_test, y_pred)

    # Store the results
    num_trees_list.append(num_trees)
    training_accuracy_list.append(training_accuracy)

    # Optionally, print out progress
    print(f"Number of Trees: {num_trees}, Accuracy: {training_accuracy:.4f}")
```
